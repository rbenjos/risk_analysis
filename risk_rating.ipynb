```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
chunk = pd.read_csv(".\data\Timestamp accuracy.csv",parse_dates=["From"])
```

```python
chunk['total error'] = chunk['Actual kWh'] - chunk['Forecast kWh']
important_cols = ['Entity','Site','Meter','From','total error','Actual kWh','Forecast kWh']
imp_chunk = chunk[important_cols]

print('done')
```

```python
new_chunk = imp_chunk.dropna()
print('done')
```

```python
by_error = new_chunk.sort_values('total error',ascending=False).reset_index(drop=True)

```

```python
by_error.to_csv(".\data\\by_error.csv")
```

```python
by_error_meter = by_error.groupby('Meter',as_index=False)['total error']
by_error_meter = by_error.loc[by_error_meter.idxmax()].sort_values('total error',ascending=False)
by_error_meter.to_csv(".\data\\by_error_meter.csv")
```

```python
pt = pd.pivot_table(data= new_chunk,values='total error',index = 'From', columns='Meter')
new_df = pd.DataFrame({'From':pt.idxmax(),'total error':pt.max()})
new_df.sort_values('total error',ascending=False)
```
